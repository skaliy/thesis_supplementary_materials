```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#colors hex:
#8fbcda
#e4acb2
#a2e0bb
#fbe4c7

#Kanskje: #3f4a55 
```

```python
df = pd.read_csv('../data/cv_conferences.csv')
```

```python
df = df.loc[df.Year > 13].reset_index(drop=True)
```

```python
df.head()
```

```python
# Plot the data using the DataFrame
fig, ax = plt.subplots()
colors = {'CVPR': '#8fbcda', 'ICCV': '#e4acb2', 'ECCV': '#a2e0bb'}
for conference, group in df.groupby('Conference'):
    group.plot(x='Year', y='Submissions', label=conference, color=colors[conference], ax=ax)

# Customize the plot's appearance
plt.xlabel('Year')
plt.ylabel('Number of submitted papers')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.title('Computer Vision and Pattern Recognition conferences')
plt.legend()
# Adding legend
plt.savefig('../figs/cv_conference.png', bbox_inches="tight") # added savefig method
```
