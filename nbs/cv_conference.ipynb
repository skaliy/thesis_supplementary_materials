```python
fig, ax = plt.subplots()
#df.plot(kind='line', color=['#8fbcda', '#e4acb2', '#a2e0bb'], ax=ax)

# Data for CVPR conference
cvpr_years = [14, 15, 16, 17, 18, 19, 20, 21, 22]
cvpr_submissions = [1807, 2123, 2145, 2620, 3303, 5160, 6656, 7015, 8161]

# Data for ICCV conference
iccv_years = [13, 15, 17, 19]
iccv_submissions = [1629, 1698, 2143, 4304]

# Data for ECCV conference
eccv_years = [14, 16, 18, 20]
eccv_submissions = [1444, 1561, 2439, 5025]

# Plotting the data
plt.plot(cvpr_years, cvpr_submissions, label='CVPR', color='#8fbcda')
plt.plot(iccv_years, iccv_submissions, label='ICCV', color='#e4acb2')
plt.plot(eccv_years, eccv_submissions, label='ECCV', color='#a2e0bb')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of submitted papers')
ax.spines['top'].set_visible(False) # removed top spine
ax.spines['right'].set_visible(False) # removed right spine
#ax.spines['bottom'].set_visible(False) # removed bottom spine
#ax.spines['left'].set_visible(False) # removed left spine

plt.title('Computer Vision and Pattern Recognition conferences')
plt.legend()
# Adding legend
plt.savefig('../figs/cv_conference.png', bbox_inches="tight") # added savefig method


# Displaying the plot
```
