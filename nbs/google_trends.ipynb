```python
df = pd.read_csv('../data/multiTimeline_news.csv', skiprows=[1,2,3,4,5])
df['Uke'] = pd.to_datetime(df['Uke'])
df.set_index('Uke', inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
plt.rcParams.update({'font.size': 11})
```

```python
fig, ax = plt.subplots()
df = df.rolling(window=4).mean()
df.plot(kind='line', color=['#8fbcda', '#e4acb2', '#a2e0bb'], ax=ax)
plt.xlabel('Date')
plt.ylabel('Search hits')
ax.spines['top'].set_visible(False) # removed top spine
ax.spines['right'].set_visible(False) # removed right spine

plt.title('Google Trends')
plt.savefig('../figs/google_trends.png', bbox_inches="tight") # added savefig method
plt.show()
```

Note that the y axis does not show the raw volume of searches, but just shows relative popularity with the max value scaled to 100.
